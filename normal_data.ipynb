```python
import os
import random
from PIL import Image

# --- נתיבי מקור (הנתיבים שסיפקת) ---
source_dirs = [
    r"C:\Users\linoy\Dataset\intel_images\seg_pred",
    r"C:\Users\linoy\Dataset\intel_images\seg_test",
    r"C:\Users\linoy\Dataset\intel_images\seg_train"
]

# --- נתיב יעד (בשרת הג'ופיטר) ---
output_normal_folder = "normal_30x30_dataset"

# --- הגדרות ---
TARGET_SIZE = (30, 30)
TOTAL_NEEDED = 1000
samples_per_dir = TOTAL_NEEDED // len(source_dirs) # כ-333 לכל תיקייה ראשית

if not os.path.exists(output_normal_folder):
    os.makedirs(output_normal_folder)

def get_all_images(directory):
    """פונקציה שמוצאת את כל התמונות בתיקייה ובתתי-תיקיות שלה"""
    valid_extensions = ('.png', '.jpg', '.jpeg')
    image_paths = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.lower().endswith(valid_extensions):
                image_paths.append(os.path.join(root, file))
    return image_paths

print("מתחיל איסוף ועיבוד תמונות נורמליות...")

overall_count = 0

for s_dir in source_dirs:
    if not os.path.exists(s_dir):
        print(f"אזהרה: התיקייה {s_dir} לא נמצאה, מדלג...")
        continue
        
    # איסוף כל התמונות מהתיקייה (כולל תתי-תיקיות לגיוון)
    all_imgs = get_all_images(s_dir)
    
    # בחירה רנדומלית של הכמות הדרושה מהתיקייה הזו
    count_to_take = min(len(all_imgs), samples_per_dir)
    selected_imgs = random.sample(all_imgs, count_to_take)
    
    dir_name = os.path.basename(s_dir)
    print(f"מעבד {count_to_take} תמונות מתוך {dir_name}...")
    
    for img_path in selected_imgs:
        try:
            img = Image.open(img_path).convert("RGB")
            
            # שינוי גודל ל-30x30
            img_resized = img.resize(TARGET_SIZE, Image.Resampling.LANCZOS)
            
            # שמירה בתיקייה המקומית
            save_name = f"normal_{overall_count}.png"
            img_resized.save(os.path.join(output_normal_folder, save_name))
            
            overall_count += 1
        except Exception:
            continue

print(f"\nסיימתי! נוצרו {overall_count} תמונות נורמליות בתיקייה: {output_normal_folder}")
```
